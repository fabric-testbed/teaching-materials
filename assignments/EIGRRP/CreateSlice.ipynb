# Create Slice for Fabric Rack
This is the slice creation for the ```EIGRPPtest``` tutorial, This is the first step of the assignment, to create the slice on fabric using fablib
 

````python
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager

fablib = fablib_manager()
                     
#fablib.show_config()

#fablib.get_image_names()
````

````python
slice_name = "EIGRPPtest"
site = "EDC"#fablib.get_random_site()
print(site)

nicmodel = "NIC_Basic"
image = "default_ubuntu_20"

cores = 1
ram = 2
disk = 10
````

````python
try:
    #Create Slice
    slice = fablib.new_slice(name=slice_name)
    
    #Server 1
    S1 = slice.add_node(name="Server1", site=site)
    S1.set_capacities(cores=cores, ram=ram, disk=disk)
    S1.set_image(image)
    S1H1Port = S1.add_component(model=nicmodel, name="s1h1Port").get_interfaces()[0] 
    S1H2Port = S1.add_component(model=nicmodel, name="s1h2Port").get_interfaces()[0] 
    S1SPort = S1.add_component(model=nicmodel, name="s1s2Port").get_interfaces()[0] 
    #s1s3Port = host1.add_component(model=nicmodel, name="s1s3Port").get_interfaces()[0] 
    
    #Server 2
    S2 = slice.add_node(name="Server2", site=site)
    S2.set_capacities(cores=cores, ram=ram, disk=disk)
    S2.set_image(image)
    S2H1Port = S2.add_component(model=nicmodel, name="s2h1Port").get_interfaces()[0] 
    S2SPort = S2.add_component(model=nicmodel, name="s2s1Port").get_interfaces()[0] 
    #S2S3Port = S2.add_component(model=nicmodel, name="s2s3Port").get_interfaces()[0] 
    
    #Server 3
    S3 = slice.add_node(name="Server3", site=site)
    S3.set_capacities(cores=cores, ram=ram, disk=disk)
    S3.set_image(image)
    S3H1Port = S3.add_component(model=nicmodel, name="s3h1Port").get_interfaces()[0] 
    S3SPort = S3.add_component(model=nicmodel, name="s3s1Port").get_interfaces()[0] 
    #S3S2Port = S3.add_component(model=nicmodel, name="s3s2Port").get_interfaces()[0] 
   
    #Host 1 Server 1
    H1S1 = slice.add_node(name="Host1_S1", site=site)
    H1S1.set_capacities(cores=cores, ram=ram, disk=disk)
    H1S1.set_image(image)
    H1S1Cp = H1S1.add_component(model=nicmodel, name="H1S1P1").get_interfaces()[0] 
     
    #Host 2 Server 1
    H2S1 = slice.add_node(name="Host2_S1", site=site)
    H2S1.set_capacities(cores=cores, ram=ram, disk=disk)
    H2S1.set_image(image)
    H2S1Cp = H2S1.add_component(model=nicmodel, name="H2_S1_P1").get_interfaces()[0] 
    
    #Host 1 Server 2
    H1S2 = slice.add_node(name="Host1_S2", site=site)
    H1S2.set_capacities(cores=cores, ram=ram, disk=disk)
    H1S2.set_image(image)
    H1S2Cp = H1S2.add_component(model=nicmodel, name="H1_S2_P1").get_interfaces()[0] 
    
    #Host 1 Server 3
    H1S3 = slice.add_node(name="Host1_S3", site=site)
    H1S3.set_capacities(cores=cores, ram=ram, disk=disk)
    H1S3.set_image(image)
    H1S3Cp = H1S3.add_component(model=nicmodel, name="H1_S3_P1").get_interfaces()[0] 
    
    
    #might merge to 1 on final version
    lh1S1 = slice.add_l2network(name="Lan1_S1H1", interfaces=[H1S1Cp, S1H1Port])
    lh2S1 = slice.add_l2network(name="Lan_S1H2", interfaces=[H2S1Cp, S1H2Port])
    lh1S2 = slice.add_l2network(name="Lan_S2", interfaces=[H1S2Cp, S2H1Port])
    lh1S3 = slice.add_l2network(name="Lan_S3", interfaces=[H1S3Cp, S3H1Port])
    lSr = slice.add_l2network(name="ServerLan", interfaces=[S1SPort, S2SPort, S3SPort])
    #Submit Slice Request
    slice.submit()
except Exception as e:
    print(f"Slice Failed: {e}")
````


### 1.2 Set up the Experiment Network

This section of the slice creation, once the nodes are active, will asign the network spaces some predefined values that we will use for the Assgnment


````python
# Setup Network
from ipaddress import ip_address, IPv4Address, IPv6Address, IPv4Network, IPv6Network

try:    
    host1_S1 = slice.get_node(name="Host1_S1") 
    host2_S1 = slice.get_node(name="Host2_S1") 
    host1_S2 = slice.get_node(name="Host1_S2") 
    host1_S3 = slice.get_node(name="Host1_S3")
    Server1 = slice.get_node(name="Server1") 
    Server2 = slice.get_node(name="Server2")
    Server3 = slice.get_node(name="Server3")
    
    subnet1 = IPv4Network("11.0.0.0/30") # S1
    subnet2 = IPv4Network("12.0.0.0/24") # S2 # 192
    subnet3 = IPv4Network("13.0.0.0/24") # S3 # 172
    #Use IP's
    #    Hosts
    h1_S1_IP = "11.0.0.3"
    h2_S1_IP = "11.0.0.4"
    h1_S2_IP = "12.0.0.2"
    h1_S3_IP = "13.0.0.2"  
    #    servers
    s1p1_IP = "11.0.1.1"   
    s1p2_IP = "11.0.0.1"   
    s1p3_IP = "11.0.0.2"    
    s2p1_IP = "12.0.1.1"
    s2p2_IP = "12.0.0.1"
    s3p1_IP = "13.0.1.1"      
    s3p2_IP = "13.0.0.1"
    
    # hosts
    h1_S1 = host1_S1.get_interface(network_name="Lan1_S1H1")
    h1_S1.ip_addr_add(addr=h1_S1_IP, subnet=subnet1) 

    h2_S1 = host2_S1.get_interface(network_name="Lan_S1H2")
    h2_S1.ip_addr_add(addr=h2_S1_IP, subnet=subnet1)
    
    h1_S2 = host1_S2.get_interface(network_name="Lan_S2")
    h1_S2.ip_addr_add(addr=h1_S2_IP, subnet=subnet2) 

    h1_S3 = host1_S3.get_interface(network_name="Lan_S3")
    h1_S3.ip_addr_add(addr=h1_S3_IP, subnet=subnet3)
    
    # servers
    s1p1 = Server1.get_interface(network_name="ServerLan")
    s1p1.ip_addr_add(addr=s1p1_IP, subnet=subnet1)    
    
    s1p2 = Server1.get_interface(network_name="Lan1_S1H1")
    s1p2.ip_addr_add(addr=s1p2_IP, subnet=subnet1)    
    
    s1p3 = Server1.get_interface(network_name="Lan_S1H2")
    s1p3.ip_addr_add(addr=s1p3_IP, subnet=subnet1)    
    
    s2p1 = Server2.get_interface(network_name="ServerLan")
    s2p1.ip_addr_add(addr=s2p1_IP, subnet=subnet2)    
    
    s2p2 = Server2.get_interface(network_name="Lan_S2")
    s2p2.ip_addr_add(addr=s2p2_IP, subnet=subnet2)
    
    s3p1 = Server3.get_interface(network_name="ServerLan")
    s3p1.ip_addr_add(addr=s3p1_IP, subnet=subnet3)    
    
    s3p2 = Server3.get_interface(network_name="Lan_S3")
    s3p2.ip_addr_add(addr=s3p2_IP, subnet=subnet3)
      
    #turn on devices
    for node in slice.get_nodes():
        for interface in node.get_interfaces():
            stdout, stderr = node.execute(f'sudo ip link set dev { interface.get_device_name()} up')
        node.execute(f'sudo sysctl -w net.ipv4.ip_forward=1')
        node.execute(f'ip a')
        
    #fix routes
    #Host routes
    print(f"New Routes AT:{host1_S1.get_name()}")
    host1_S1.execute(f'sudo ip route flush dev {h1_S1.get_device_name()}')
    host1_S1.execute(f'sudo ip route add {s1p2_IP} dev {h1_S1.get_device_name()}')
    host1_S1.execute("ip route")   
    print(f"New Routes AT:{host2_S1.get_name()}")
    host2_S1.execute(f'sudo ip route flush dev {h2_S1.get_device_name()}')
    host2_S1.execute(f'sudo ip route add {s1p3_IP} dev {h2_S1.get_device_name()}')
    host2_S1.execute("ip route")  
    print(f"New Routes AT:{host1_S2.get_name()}")
    host1_S2.execute(f'sudo ip route flush dev {h1_S2.get_device_name()}')
    host1_S2.execute(f'sudo ip route add {s2p2_IP} dev {h1_S2.get_device_name()}')
    host1_S2.execute("ip route") 
    print(f"New Routes AT:{host1_S3.get_name()}")
    host1_S3.execute(f'sudo ip route flush dev {h1_S3.get_device_name()}')
    host1_S3.execute(f'sudo ip route add {s3p2_IP} dev {h1_S3.get_device_name()}')
    host1_S3.execute("ip route")    

    #Server Routes
    print(f"New Routes AT:{Server1.get_name()}")
    Server1.execute(f'sudo ip route flush dev {s1p1.get_device_name()}')
    Server1.execute(f'sudo ip route flush dev {s1p2.get_device_name()}')
    Server1.execute(f'sudo ip route flush dev {s1p3.get_device_name()}')
    Server1.execute(f'sudo ip route add {subnet2} dev {s1p1.get_device_name()}')
    Server1.execute(f'sudo ip route add {subnet3} dev {s1p1.get_device_name()}')
    Server1.execute(f'sudo ip route add {h1_S1_IP} dev {s1p2.get_device_name()}')
    Server1.execute(f'sudo ip route add {h2_S1_IP} dev {s1p3.get_device_name()}')
    Server1.execute("ip route")
    print(f"New Routes AT:{Server2.get_name()}")
    Server2.execute(f'sudo ip route flush dev {s2p1.get_device_name()}')
    Server2.execute(f'sudo ip route flush dev {s2p2.get_device_name()}')
    Server2.execute(f'sudo ip route add {subnet1} dev {s2p1.get_device_name()}')
    Server2.execute(f'sudo ip route add {subnet3} dev {s2p1.get_device_name()}')
    Server2.execute(f'sudo ip route add {h1_S2_IP} dev {s2p2.get_device_name()}')
    Server2.execute("ip route")   
    print(f"New Routes AT:{Server3.get_name()}")
    Server3.execute(f'sudo ip route flush dev {s3p1.get_device_name()}')
    Server3.execute(f'sudo ip route flush dev {s3p2.get_device_name()}')
    Server3.execute(f'sudo ip route add {subnet2} dev {s3p1.get_device_name()}')
    Server3.execute(f'sudo ip route add {subnet1} dev {s3p1.get_device_name()}')
    Server3.execute(f'sudo ip route add {h1_S3_IP} dev {s3p2.get_device_name()}')
    Server3.execute("ip route")  
    
    
except Exception as e:
        print(f"Exception: {e}")
````
